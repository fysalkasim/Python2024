```python
def is_palindrome(s):
    og = list(s)
    rev = og.copy()
    rev.reverse()
    if og == rev:
        print("its is a palindrome")
```

```python
def is_palindrome(s):
    if s == s[::-1]:
        print("its is a palindrome")
```

```python
is_palindrome("MalayalaM")
```

```python
def is_palindrome(s):
    for i in range(len(s)):
        if s[i] != s[-i]:
            print("Not a palindrome")
            print(i)
        else:
            print("Its is a palindrome")
            
```

```python
is_palindrome("MalayalaM")
```
